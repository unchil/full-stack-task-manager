```kotlin
%useLatestDescriptors
%use kandy
%use dataframe
```

```kotlin
var df = DataFrame.readJson("http://localhost:7788/nifs/stat")
```

```kotlin
df.describe()
```

```kotlin
var df2 = df.convert { tmp_min and  tmp_max and  tmp_avg }.with { it.toFloat() }
df2.describe()
```

```kotlin
df2.filter { gru_nam == "서해" }.plot{

    layout {
        title = "일별 해수 온도 변화"
        size = 2000 to 800
    }

    x("obs_datetime"){
        axis.name = "관측일시"

    }

    y {
        axis {
            limits = tmp_min.min().toInt().minus(1)..tmp_max.max().toInt().plus(1)
            name = "Temperature"
        }
    }

    line {
        y(tmp_avg)
        color("sta_nam_kor"){
            legend.name = "관측소명"
        }
        type = LineType.DOTTED
        width = 0.6
    }

    ribbon {
        yMin("tmp_min")
        yMax("tmp_max")
        fillColor("sta_nam_kor") {
            legend.name = "관측소명"
        }
        alpha = 0.6
        borderLine.width = 0.0
    }
}
```
